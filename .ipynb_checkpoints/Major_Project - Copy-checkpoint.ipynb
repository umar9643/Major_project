```python
def treat_price(x):
    # If already float (numeric), return as is
    if isinstance(x, float) or isinstance(x, int):
        return x
    
    # If x is not a list/tuple, return NaN
    if not isinstance(x, (list, tuple)):
        return None
    
    # If the list length is less than 2, return NaN
    if len(x) < 2:
        return None
    
    # Handle "Lac" values
    if x[1] == "Lac":
        return round(float(x[0]) / 100, 2)
    else:  # Assume "Cr" or similar
        return round(float(x[0]), 2)
```

```python
#price_per_sqft
df["price_per_sqft"].value_counts()
```

```python
df["price_per_sqft"] = (
    df["price_per_sqft"]
    .str.split("/")
    .str.get(0)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

# Convert safely → invalid strings become NaN
df["price_per_sqft"] = pd.to_numeric(df["price_per_sqft"], errors="coerce")

# Drop those invalid rows
df = df.dropna(subset=["price_per_sqft"])
```
